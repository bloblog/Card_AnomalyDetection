## 모듈 & 데이터 임포트 

```python
import pandas as pd
import numpy as np
import warnings
import random
warnings.filterwarnings("ignore")
```

```python
from sklearn.covariance import EllipticEnvelope
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import classification_report
```

```python
train = pd.read_csv('train.csv')
val = pd.read_csv('val.csv')
test = pd.read_csv('test.csv')
```

```python
train_df = train.drop(columns=['ID'])
val_df = val.drop(columns=['ID'])

train_all = train_df
val_all = val_df.drop(columns=['Class'])
val_y = val_df['Class']
test_vari = test.drop(columns=['ID'])
```

```python
from sklearn.preprocessing import StandardScaler  # 표준화 
std = StandardScaler()

train_std = std.fit_transform(train_all) 
val_std = std.fit_transform(val_all) 
test_std = std.fit_transform(test_vari)
```

```python
# 레이블 변경 함수
def get_pred_label(model_pred):
  model_pred = np.where(model_pred==1, 0, model_pred)
  model_pred = np.where(model_pred==-1, 1, model_pred)
  return model_pred
```

```python
EE_model = EllipticEnvelope(support_fraction = 0.999, contamination = 0.00112, random_state = 42)
EE_model.fit(train_std)
```

```python
test_EE_pred = EE_model.predict(test_vari)
test_EE_pred1 = get_pred_label(test_EE_pred)
```

```python
submission = pd.read_csv('sample_submission.csv')
submission.head()
```

```python
# 테스트 데이터 레이블 입력
submission['Class'] = test_EE_pred1

submission
```

```python
submission.to_csv('submission.csv',index=False)
```
